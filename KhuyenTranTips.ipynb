## AB Testing with leAB module

```python
from leab import before
# What is the number of samples needed per
# variation to detect a 1% result difference
# in a population with a 15% conversion rate?
ab_test = before.leSample(
    conversion_rate=15, min_detectable_effect=1
)
ab_test.get_size_per_variation()
```

```python
from leab import after, leDataset

# Import sample data from A and B
data = leDataset.SampleLeSuccess()

# Compose the successes between the two groups
ab_test = after.leSucess(
    data.A, data.B, confidence_level=0.95
)
# Get the conclusion on the test
ab_test.get_verdict()
```

## Dataclasses provide concise syntax for data-holding classes.
Pair it with 'pyserde' allows you to serialize and deserialize dataclass objects.

```python
from dataclasses import dataclass
from serde import serde
from serde.json import from_json, to_json

@serde
@dataclass
class User:
  name: str
  age: int

user1 = User(name="user1", age=20)

# serialize
print(to_json(user1))

# deserialize
print(from_json(User, '{"name" : "user1", "age":20}'))
```

## To ensure that your application logs an error under
## a specific condition, use the built-in called caplog in pytest

```python
from logging import getLogger

logger = getLogger(__name__)

def divide(num1: float, num2: float) --> float:
  if num2 == 0:
    logger.error(f"Can't divide {num1} by 0")
  else:
    logger.info(f"Divide {num1} by {num2}")
    return num1/num2

def test_divide_by_0(caplog):
  divide(1,0)
  assert "Can't divide 1 by 0" in caplog.text
```
